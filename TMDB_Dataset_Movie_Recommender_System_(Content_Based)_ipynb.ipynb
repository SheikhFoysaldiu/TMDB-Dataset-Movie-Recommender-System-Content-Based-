```python
import ast
```

```python
def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i["name"])
  return L;
```

```python
# convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
def convertTopCast(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter!=3:
      L.append(i["name"])
      counter+=1
    else:
      break
  return L;
```

```python
movies['cast'] = movies['cast'].apply(convertTopCast)
```

```python
movies.head(1)
```

```python
movies['crew'][0]
```

```python
def convertDirectCrew(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i["name"])
      break
  return L;
```

```python
movies['crew'] = movies['crew'].apply(convertDirectCrew)
```

**Applying Lemmatization**

```python
import nltk
nltk.download("all")
```

```python
from nltk.stem import 	WordNetLemmatizer
from nltk.tokenize import word_tokenize
lemmatizer = WordNetLemmatizer()
```

```python
def lemma(text):
  y = []
  for i in word_tokenize(text):
    y.append(lemmatizer.lemmatize(i))
  return " ".join(y)
```

```python
print("jamescameron jamescameron cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d action adventure fantasy sciencefiction in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization.")
lemma("jamescameron jamescameron cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d action adventure fantasy sciencefiction in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization.")
```

```python
df['tags'] = df['tags'].apply(lemma)
```
